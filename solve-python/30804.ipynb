```python
def solution(N, S):
  if N <= 2: return N
  l, r = 0, 1
  ans = 2
  temp_ans = 2
  pair = [S[l], S[r]]

  while r < N - 1:
    r += 1
    temp_ans += 1
    if S[r] not in pair:
      temp_ans = 2
      l = r - 1
      while S[l - 1] == S[l]:
        l -= 1
        temp_ans += 1
      pair = [S[l], S[r]]
    ans = max(ans, temp_ans)
  return ans

N = int(input())
S = list(map(int, input().split()))

print(solution(N, S))
```
